Looks like df_data has all unique IDs. df_symptoms has a ton of duplicates. 
Clean up df_vax first since it has the least duplicates

```python
# Get all the IDs that are duplicates
df_vax['VAERS_ID'].value_counts()[df_vax['VAERS_ID'].value_counts() > 1]
```

```python
# Look at one ID and see what's going on here
df_vax[df_vax['VAERS_ID'] == 1022135]
```

So it looks like we have multiple vaccines for each person. We are just focused on COVID here, so let's pull all those.

```python
clean_vax = df_vax[df_vax['VAX_TYPE'] == 'COVID19']
clean_vax
```

```python
# Get all the IDs that are duplicates
dupes = clean_vax['VAERS_ID'].value_counts()[clean_vax['VAERS_ID'].value_counts() > 2]
dupes
```

```python
# Look at one ID and see what's going on here
clean_vax[clean_vax['VAERS_ID'] == 931569]
```

Looks like our remaining duplicate IDs are for 1st and 2nd dose when there's two duplicates. If there's more than 2 duplicates, it looks like errors in data entry. We will remove anything with more than two duplicates.

```python
dupes.index
```

```python
clean_vax[clean_vax['VAERS_ID'].isin(dupes.index)]
```

Actually, we can salvage some of these rows in case the ID's counterpart is in the other dataframes.

```python
salvage = clean_vax[clean_vax['VAERS_ID'].isin(dupes.index)].loc[[13245, 15537, 33516, 33517]]
salvage
```

```python
clean_vax2 = clean_vax[~clean_vax['VAERS_ID'].isin(dupes.index)].append(salvage)
clean_vax2
```

```python
clean_vax2['VAERS_ID'].value_counts()[clean_vax2['VAERS_ID'].value_counts() > 1]
```

```python
df_covid = pd.merge(df_data, clean_vax2, on=['VAERS_ID'])
df_covid
```
