```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
```

```python
df_symptoms = pd.read_csv(io.BytesIO(uploaded['2021VAERSSYMPTOMS.csv']))
```

```python
df_vax = pd.read_csv(io.BytesIO(uploaded['2021VAERSVAX.csv']))
```

```python
df_data = pd.read_csv(io.BytesIO(uploaded['2021VAERSDATA.csv']), encoding="latin1")
```

```python
df_half = pd.merge(df_data, df_symptoms)
df_covid = pd.merge(df_half, df_vax)
```
