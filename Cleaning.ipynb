We took all the duplicates in the symptoms data and collapsed the rows to create 10 columns for symptoms instead of 2 rows of symptoms for the same person.

Now we want to collapse df_final's duplicate IDs into one row to make the final merge easier. We'll group by VAERS_ID, and then for each group, if there's two rows in the group, we take the row with the least NaNs.

```python
blah = df_final['VAERS_ID'].value_counts()[df_final['VAERS_ID'].value_counts() == 2]
test = df_final[df_final['VAERS_ID'].isin(blah.index)]
```

```python
# If there is more than 1 row, return the row with the least nans
def f(group):
    if len(group) == 2:
        if group.iloc[0].isna().sum() > group.iloc[1].isna().sum():
            return group.iloc[1]       
    return group.iloc[0]
```

```python
test.iloc[0:8]
```

```python
test.iloc[0:8].groupby("VAERS_ID").apply(f)
```

Now we apply our function to the real dataset.

```python
df_final
```

```python
df_clean = df_final.groupby("VAERS_ID").apply(f)
df_clean
```

```python
df_clean.reset_index(drop=True, inplace=True)
df_clean
```
